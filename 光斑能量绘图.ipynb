```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# 生成100*100的A矩阵，每个元素都是1
A = np.ones((100, 100))
# 生成160*160的B矩阵，每个元素都是1
B = np.ones((160, 160))

# 遍历A矩阵的每个元素，A矩阵对应的B矩阵中心的100*100的矩阵里的元素周围半径为30的元素都加1
for i in range(100):
    for j in range(100):
        for x in range(-30, 30):
            for y in range(-30, 30):
                if x**2 + y**2 <= 30**2:
                    B[i+30+x][j+30+y] += 1
```

```python
print(B)
```

```python
# 以B矩阵为像素点，画出热力图
import seaborn as sns
sns.heatmap(B, cmap='inferno')
plt.show()
```
